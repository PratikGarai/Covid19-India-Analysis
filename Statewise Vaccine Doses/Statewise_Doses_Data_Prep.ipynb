```python
import pandas as pd
```

```python
df = pd.read_csv("vaccine_doses_statewise.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
states = df['State'].unique()
states
```

```python
cols = df.columns
cols
```

```python
df2 = df.drop(['State'],axis=1).transpose()
df2.columns = states
df2.head()
```

```python
df2[states[0]].values
```

```python
indexes = df2.index
indexes
```

```python
ls = []
for i in states :
    data = df2[i]
    for j,k in zip(indexes, data) :
        ls.append([i, j, k])
```

```python
df3 = pd.DataFrame(ls)
df3
```

```python
df3.info()
```

```python
df3.columns = ['State', 'Date', 'Doses']
df3.head()
```

```python
df3.to_csv("statewise_doses_data_prep.csv", index=False)
```
